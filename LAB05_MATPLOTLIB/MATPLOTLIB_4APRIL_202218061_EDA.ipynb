#USING LINEAR INTERPOLATION

```python
def linear_interpolation(column):
    index = np.arange(len(column))
    is_nan = np.isnan(column)
    interp = interp1d(index[~is_nan], column[~is_nan], kind='linear')
    column[is_nan] = interp(index[is_nan])
    return column
```

```python
def spline_interpolation(column):
    index = np.arange(len(column))
    is_nan = np.isnan(column)
    interp = interp1d(index[~is_nan], column[~is_nan], kind='cubic')
    column[is_nan] = interp(index[is_nan])
    return column
```
